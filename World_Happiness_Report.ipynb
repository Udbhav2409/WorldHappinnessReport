## **1. Downloading and Loading the Dataset**
This section downloads the dataset from Kaggle, loads it into a Pandas DataFrame, and displays the first few rows.


```python
import pandas as pd
import os
import kagglehub

dataset_path = kagglehub.dataset_download("samithsachidanandan/world-happiness-report-2020-2024")

print("Dataset files:", os.listdir(dataset_path))

df = pd.read_csv(os.path.join(dataset_path, "2020.csv"))

df.head()
```

## **2. Understanding the Dataset**
This section provides basic information about the dataset, including column names, data types, and missing values.


```python
df.info()

df.isnull().sum()
```

## **3. Distribution of Happiness Scores**
A histogram is used to visualize the distribution of happiness scores across different countries.


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.histplot(df["Happiness score"], kde=True, color="skyblue")
plt.title("Distribution of Happiness Scores")
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.show()
```

## **4. Happiness Score vs GDP per Capita**
A scatter plot is used to analyze the relationship between a country's happiness score and its GDP per capita.


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Economy (GDP per Capita)\t"], y=df["Happiness score"], color="green")
plt.title("Happiness Score vs GDP per Capita")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.show()
```

## **5. Correlation Matrix**
A heatmap is generated to visualize the correlation between different numerical variables.


```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Happiness Factors")
plt.show()
```

## **6. Average Happiness Score by Region**
If available, this section groups the dataset by regions and calculates the average happiness score for each.


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Social support"], y=df["Happiness score"], color="purple")
plt.title("Impact of Social Support on Happiness Score")
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")
plt.show()
```

## **7. Impact of Social Support on Happiness**
A scatter plot is used to analyze how social support influences the happiness score.


```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Healthy life expectancy"], y=df["Happiness score"], color="red")
plt.title("Impact of Healthy Life Expectancy on Happiness Score")
plt.xlabel("Healthy Life Expectancy")
plt.ylabel("Happiness Score")
plt.show()
```
